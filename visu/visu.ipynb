```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charger le fichier CSV avec pandas
df = pd.read_csv('../data/farms_train.csv', delimiter=';', decimal=',')

# Sélectionner les colonnes B à G (en supposant que les colonnes sont étiquetées A, B, C,...)
# Si les colonnes sont étiquetées de façon numérique, vous pouvez les sélectionner par indices.
df_subset = df.iloc[:, 1:7]  # Sélectionne les colonnes B à G (index 1 à 6)

# Calculer la matrice de corrélation
corr_matrix = df_subset.corr()

# Visualiser la matrice de corrélation
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matrice de Corrélation entre les variables')
plt.show()
```

Histogrammes individuels des variables

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math

# Charger le fichier CSV avec pandas
df = pd.read_csv('../data/farms_train.csv', delimiter=';', decimal=',')

# Enlever la première colonne
df = df.drop(columns=df.columns[0])

# Calculer le nombre de colonnes pour organiser la grille
num_cols = len(df.columns)
cols = 3
rows = math.ceil(num_cols / cols)

# Créer les sous-graphiques
fig, axes = plt.subplots(rows, cols, figsize=(15, rows * 3))

for i, col in enumerate(df.columns):
    row = i // cols
    col_pos = i % cols
    sns.histplot(df[col], kde=True, ax=axes[row, col_pos])
    axes[row, col_pos].set_title(f'Distribution de {col}')
    axes[row, col_pos].set_xlabel(col)
    axes[row, col_pos].set_ylabel('Fréquence')

# Supprimer les axes vides si le nombre de graphiques est inférieur à la grille
for j in range(i + 1, rows * cols):
    fig.delaxes(axes.flatten()[j])

plt.tight_layout()
plt.show()
```
