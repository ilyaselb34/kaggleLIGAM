```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charger le fichier CSV avec pandas
df = pd.read_csv('data/farms_train.csv', delimiter=';', decimal=',')

# Slectionner les colonnes B  G (en supposant que les colonnes sont tiquetes A, B, C,...)
# Si les colonnes sont tiquetes de faon numrique, vous pouvez les slectionner par indices.
df_subset = df.iloc[:, 1:7]  # Slectionne les colonnes B  G (index 1  6)

# Calculer la matrice de corrlation
corr_matrix = df_subset.corr()

# Visualiser la matrice de corrlation
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matrice de Corrlation entre les colonnes B  G')
plt.show()
```
