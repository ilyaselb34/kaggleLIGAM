```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charger le fichier CSV avec pandas
df = pd.read_csv('../data/farms_train.csv', delimiter=';', decimal=',')

# Sélectionner les colonnes B à G (en supposant que les colonnes sont étiquetées A, B, C,...)
# Si les colonnes sont étiquetées de façon numérique, vous pouvez les sélectionner par indices.
df_subset = df.iloc[:, 1:7]  # Sélectionne les colonnes B à G (index 1 à 6)

# Calculer la matrice de corrélation
corr_matrix = df_subset.corr()

# Visualiser la matrice de corrélation
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matrice de Corrélation entre les variables')
plt.show()
```

Histogrammes individuels des variables

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charger le fichier CSV avec pandas
df = pd.read_csv('../data/farms_train.csv', delimiter=';', decimal=',')


for col in df.columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution de {col}')
    plt.xlabel(col)
    plt.ylabel('Fréquence')
    plt.show()
```

Scatter plots pour les paires de variables corrélées

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../data/farms_train.csv', delimiter=';', decimal=',')

sns.scatterplot(x=df['R2'], y=df['R32'])
plt.title('Relation entre R2 et R32')
plt.xlabel('R2')
plt.ylabel('R32')
plt.show()

sns.scatterplot(x=df['R7'], y=df['R32'])
plt.title('Relation entre R7 et R32')
plt.xlabel('R7')
plt.ylabel('R32')
plt.show()


```
