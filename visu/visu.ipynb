```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charger le fichier CSV avec pandas
df = pd.read_csv('../data/farms_train.csv', delimiter=';', decimal=',')

# Slectionner les colonnes B  G (en supposant que les colonnes sont tiquetes A, B, C,...)
# Si les colonnes sont tiquetes de faon numrique, vous pouvez les slectionner par indices.
df_subset = df.iloc[:, 1:7]  # Slectionne les colonnes B  G (index 1  6)

# Calculer la matrice de corrlation
corr_matrix = df_subset.corr()

# Visualiser la matrice de corrlation
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matrice de Corrlation entre les variables')
plt.show()
```

Histogrammes individuels des variables

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
# Charger le fichier CSV avec pandas
df = pd.read_csv('../data/farms_train.csv', delimiter=';', decimal=',')



num_cols = len(df.columns)

cols = 3
rows = math.ceil(num_cols / cols)

fig, axes = plt.subplots(rows, cols, figsize=(15, rows * 3))

for i, col in enumerate(df.columns):
    row = i // cols
    col_pos = i % cols
    sns.histplot(df[col], kde=True, ax=axes[row, col_pos])
    axes[row, col_pos].set_title(f'Distribution de {col}')
    axes[row, col_pos].set_xlabel(col)
    axes[row, col_pos].set_ylabel('Frquence')

for j in range(i + 1, rows * cols):
    fig.delaxes(axes.flatten()[j])

plt.tight_layout()
plt.show()


```
